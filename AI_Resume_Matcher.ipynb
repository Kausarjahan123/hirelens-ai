```python
import pdfplumber

text = ""

with pdfplumber.open("ShortResumeK.pdf") as pdf:
    for page in pdf.pages:
        text += page.extract_text() or ""

print(text[:2000])
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model Loaded")
```

```python
job_description = """
We are looking for a Machine Learning Engineer with experience in Python,
Scikit-learn, NLP, and building machine learning models. Knowledge of APIs
and deployment is a plus.
"""
```

```python
resume_embedding = model.encode(text)
job_embedding = model.encode(job_description)
```

```python
score = cosine_similarity([resume_embedding], [job_embedding])[0][0]

print("Match Score:", round(score * 100, 2), "%")
```
